```python
import requests
import json
import pandas as pd
```

```python
api_url = 'https://shopee.vn/api/v4/search/search_items?by=relevancy&limit=60&match_id=11035567&newest=0&order=desc&page_type=search&scenario=PAGE_OTHERS&version=2'
image_url = 'https://cf.shopee.vn/file/'
```

```python
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.54 Safari/537.36'}
```

```python
response = requests.get(api_url, headers = headers)
```

```python
shopee_data = response.text
```

```python
shopee_dict = json.loads(shopee_data)
shopee_dict['items']
```

```python
shopee_datas = shopee_dict['items']
product_list = []

for product_detail in shopee_datas:
    image_ids_list = []
    item_id = product_detail.get('item_basic').get('itemid')
    name = product_detail.get('item_basic').get('name')
    price = product_detail.get('item_basic').get('price')
    quantity_sold = product_detail.get('item_basic').get('sold')
    rating = product_detail.get('item_basic').get('item_rating').get('rating_star')
    image_ids = product_detail.get('item_basic').get('images')
    
    if quantity_sold:
        sold = quantity_sold
    else:
        sold = 'n/a'
    
    if not rating:
        rating = 'n/a'
 
    for i in range(3):
        try:
            image_ids_list.append(image_ids[i])
        except:
            image_ids_list.append('')

    image_url_1 = image_url + image_ids_list[0]
    image_url_2 = image_url + image_ids_list[1]
    image_url_3 = image_url + image_ids_list[2]
    
    product_dict = {'item_id': item_id,
                   'name': name,
                   'price': price,
                   'sold': sold,
                   'rating': rating,
                   'image_url_1': image_url_1,
                   'image_url_2': image_url_2,
                   'image_url_3': image_url_3}
    
    product_list.append(product_dict)
```

```python
df = pd.DataFrame(product_list)
df.to_csv('shopee_list.csv', index = False, encoding = 'utf-8-sig')
```
