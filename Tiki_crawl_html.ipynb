```python
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup as bs
import re
```

```python
html_url = 'https://tiki.vn/sach-ky-nang-song/c871?__s=ssdp&ssdp_block_code=featured_categories'
```

```python
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.54 Safari/537.36'}
```

```python
response = requests.get(html_url, headers = headers)
soup = bs(response.content, "html.parser")
print(soup.prettify())
```

```python
soup.find_all("div", attrs = {"data-view-id" : "product_list_container"})
```

```python
a_tags = soup.find_all("a", attrs = {"class": "product-item"})
a_tags_dict = {}
a_tags_list = []
```

```python
for tags in a_tags:
    name = tags.find(class_ = "name").text
    try:
        discount = tags.find(class_ = "price-discount__discount").text
    except:
        discount = 0
    price = tags.find(class_ = "price-discount__price").text
    a_tags_dict = {"name": name,
                  "discount": discount,
                  "price": price}
    a_tags_list.append(a_tags_dict)
```

```python
df = pd.DataFrame(a_tags_list)
df
```

```python
df['discount'] = df["discount"].str.replace(r'[-%]', '', regex=True)
df['price'] = df["price"].str.replace(r'[.₫]', '', regex=True)
```

```python
df.to_csv("tiki_data_fn.csv", index = False, encoding = "utf-8-sig")
```
